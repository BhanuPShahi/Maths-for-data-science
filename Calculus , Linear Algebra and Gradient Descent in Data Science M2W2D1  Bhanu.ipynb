##**Calculus**

Calculus is the branch of Mathematics in which we are dealing with rate of change of quantities, areas, length and their volumes. It is an intrinsic field of maths and especially in many machine learning algorithms that you cannot think of skipping this to learn the essence of Data Science.

It is divided into two parts :


*   Differential Calculus
*   Integral Calculus

**Differential Calculus** is about cutting something into small pieces to find how it changes while **Integral Calculus** is about integrating (joining) those small pieces to find how much they are.

We need the understanding of both to implement ML algorithms in data science.
 

Machine Learning use cases in Calculus

*  Numerical Optimizations
*  Gradient Computaions
*  Probability density functions
*  Variational inference and related techniques

##**Linear Algebra**

Linear Algebra plays a vital and important role in the data science field. In linear algebra we are dealing with vectors , matrices and their various operations. A good understanding of these concepts helps in better understanding and visualisation of machine learning algorithms , various transformations, image recognition and dimensionality reduction algorithms etc.

With an understanding of Linear Algebra, we will be able to develop a better intuition for machine learning and deep learning algorithms and not treat them as black boxes. This would allow us to choose proper hyperparameters and develop a better model.

We would also be able to code algorithms from scratch and make our own variations to them as well. Consider linear algebra as the key to unlock a whole new world.

**Vectors**

Abstractly, vectors are objects that can be added together (to form new vectors) and that can be multiplied by scalars (i.e., numbers), also to form new vectors.

e.g. if we have the heights, weights, and ages of a large number of people, we can treat our data as three-dimensional vectors (height, weight, age). If we are teaching a class with four exams, we can treat student grades as four-dimensional vectors (exam1, exam2, exam3, exam4).

Vectors and matrices are a key part of everything we do in the Data Science world. For example, when were training a model, were fitting a matrix of features, when we look at the distance between elements in our data, were usually finding some sort of geometric distance between vectors of features and when were using a dimensionality reduction technique, were usually just finding a condensed way of representing a set of vectors without loosing the underlying relationships between them. Thats why we need to start by understanding what these elements are.

We need the understanding of following operations on vectors.

*   Vector Subtraction
*   Scaling of vectors
*   Dot product
*   Cross product
*   Vector Projections
*   Orthogonality

**Matrices**

A matrix is a two-dimensional collection of numbers. We will represent matrices as lists of lists, with each inner list having the same size and representing a row of the matrix. If A is a matrix, then A[i][j] is the element in the ith row and the jth column.

In Machine Learning matrix operations are everywhere. A matrix is no more than a two-dimensional array with one ore more columns and with one or more rows. Sounds familiar? Does it sounds perhaps like a Pandas DataFrame? Bingo! And guess what: every time were performing an addition, subtraction or multiplication on any of our features, we are usually performing a matrix operation. Moreover, most of the algorithms we use the most somehow use matrix operations under the hood. Neural networks for example are a chain of matrix operations happening one after another. 

We need the understanding of following topics in  matrices.

*  Theory of Matrices
*  Types of matrices
*  Determinant of a matrix
*  Inverse of a matrix

**Eigenvectors and eigenvalues**

Eigenvectors and eigenvalues are the final milestone in our Linear Algebra journey. Dimensionality reductions algorithms, just as Principal Component Analysis rely on these concepts heavily, and some real world tools such as the Google Search algorithm use them as a key part of their logic.

##**Gradient Descent**

**Uses of Gradient Descent**

Gradient Descent is an optimization algorithm for finding a local minimum of a differentiable function. Gradient descent is simply used to find the values of a function's parameters (coefficients) that minimize a cost function as far as possible.

A gradient measures how much the output of a function changes if we change the inputs a little bit. We can also think of a gradient as the slope of a function. The higher the gradient, the steeper the slope and the faster a model can learn. But if the slope is zero, the model stops learning. In mathematical terms, a gradient is a partial derivative with respect to its inputs.

In our machine learning models we have a Loss Function which is a function of actual values and predicted values. 

**L.F. = g (A,P)**

Here our aim is to minimize this loss function for making best machine learning model. So minimizing this, basically means getting to the lowest error value possible or increasing the accuracy of the model. In short, We increase the accuracy by iterating over a training data set of our model.

Imagine there is a hill and a blindfolded man who wants to come to the bottom of the hill with the fewest steps along the way as possible.
We can think of gradient descent as hiking down of that blindfolded man to the bottom of the valley.

GradDescent.jpg

Gradient descent is a first order (requires first derivative) optimization algorithm used to minimize a given objective function. The algorithm depends on the initial value and the learning rate. The idea behind gradient descent is that it takes a seqeuence of steps to go from the starting position to a local minimum. An initial value that is somewhat close to the global minimum is ideal. Too small a learning rate makes it more likely for the algorithm to get stuck at a local minimum. Whereas too high a learning rate makes it more likely that the algorithm will skip over the global minimum. The gradient descent can be applied to any first differentiable function.

To calculate and model, gradient descent requires calculus and now we can see importance of calculus in machine learning.
